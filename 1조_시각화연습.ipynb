```python
import pandas as pd
data = pd.read_csv('/content/sample_data/airdata.csv')
```

```python
data.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import datetime
april = data[data["날짜"].str.contains("2023-04")]
august = data[data["날짜"].str.contains("2023-08")]
december = data[data["날짜"].str.contains("2023-12")]
```

```python
sns.boxplot(april['PM10'])
```

```python
df = april.pivot(index='날짜',columns='시간',values='PM10')
sns.heatmap(df, annot=False, cmap='spring', center=100)
plt.show()
```

```python

df1 = april.pivot(index='날짜',columns='시간',values='PM10')
df2 = august.pivot(index='날짜',columns='시간',values='PM10')
df3 = december.pivot(index='날짜',columns='시간',values='PM10')
df4 = april.pivot(index='날짜',columns='시간',values='PM2.5')
df5 = august.pivot(index='날짜',columns='시간',values='PM2.5')
df6 = december.pivot(index='날짜',columns='시간',values='PM2.5')
df7 = april.pivot(index='날짜',columns='시간',values='오존')
df8 = august.pivot(index='날짜',columns='시간',values='오존')
df9 = december.pivot(index='날짜',columns='시간',values='오존')
df10 = april.pivot(index='날짜',columns='시간',values='이산화질소')
df11 = august.pivot(index='날짜',columns='시간',values='이산화질소')
df12 = december.pivot(index='날짜',columns='시간',values='이산화질소')
df13 = april.pivot(index='날짜',columns='시간',values='아황산가스')
df14 = august.pivot(index='날짜',columns='시간',values='아황산가스')
df15 = december.pivot(index='날짜',columns='시간',values='아황산가스')
df16 = april.pivot(index='날짜',columns='시간',values='일산화탄소')
df17 = august.pivot(index='날짜',columns='시간',values='일산화탄소')
df18 = december.pivot(index='날짜',columns='시간',values='일산화탄소')

plt.subplot(6,3,1)
sns.heatmap(df1, annot=False, cmap='spring', center=100)
plt.title('APRIL PM10')
plt.subplot(6,3,2)
sns.heatmap(df2, annot=False, cmap='summer', center=100)
plt.title('AUGUST PM10')
plt.subplot(6,3,3)
sns.heatmap(df3, annot=False, cmap='winter',center = 100)
plt.title('DECEMBER PM10')
plt.subplot(6,3,4)
sns.heatmap(df4, annot=False, cmap='spring', center=50)
plt.title('APRIL PM2.5')
plt.subplot(6,3,5)
sns.heatmap(df5, annot=False, cmap='summer', center=50)
plt.title('AUGUST PM2.5')
plt.subplot(6,3,6)
sns.heatmap(df6, annot=False, cmap='winter', center=50)
plt.title('DECEMBER PM2.5')
plt.subplot(6,3,7)
sns.heatmap(df7, annot=False, cmap='spring', center=0.1)
plt.title('APRIL OZONE')
plt.subplot(6,3,8)
sns.heatmap(df8, annot=False, cmap='summer', center=0.1)
plt.title('AUGUST OZONE')
plt.subplot(6,3,9)
sns.heatmap(df9, annot=False, cmap='winter', center=0.1)
plt.title('DECEMBER OZONE')
plt.subplot(6,3,10)
sns.heatmap(df10, annot=False, cmap='spring', center=0.6)
plt.title('APRIL NO2')
plt.subplot(6,3,11)
sns.heatmap(df11, annot=False, cmap='summer', center=0.6)
plt.title('AUGUST NO2')
plt.subplot(6,3,12)
sns.heatmap(df12, annot=False, cmap='winter', center=0.6)
plt.title('DECEMBER NO2')
plt.subplot(6,3,13)
sns.heatmap(df13, annot=False, cmap='spring', center=0.02)
plt.title('APRIL SO2')
plt.subplot(6,3,14)
sns.heatmap(df14, annot=False, cmap='summer', center=0.02)
plt.title('AUGUST SO2')
plt.subplot(6,3,15)
sns.heatmap(df15, annot=False, cmap='winter', center=0.02)
plt.title('DECEMBER SO2')
plt.subplot(6,3,16)
sns.heatmap(df16, annot=False, cmap='spring', center=0.6)
plt.title('APRIL CO')
plt.subplot(6,3,17)
sns.heatmap(df17, annot=False, cmap='summer', center=0.6)
plt.title('AUGUST CO')
plt.subplot(6,3,18)
sns.heatmap(df18, annot=False, cmap='winter', center=0.6)
plt.title('DECEMBER CO')
fig = plt.gcf()
fig.set_size_inches(36,36)
plt.show()
```
