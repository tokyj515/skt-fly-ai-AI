# 인사이드아웃 댓글 크롤링

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import matplotlib.font_manager as fm
import warnings

warnings.filterwarnings(action='ignore')

# 폰트 설정
path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
font_name = fm.FontProperties(fname=path, size=10).get_name()
plt.rcParams['font.family'] = font_name

# 크롤링
url = 'http://cine21.com/movie/info/?movie_id=43423'
res = requests.get(url)
soup = bs(res.content, 'html.parser')

# 전문가 평가
expert = soup.select('div.expert_rating_area span.comment')

expert_comments = ''
for comment in expert:
    expert_comments += comment.get_text() + ' '

# WordCloud 생성
wordcloud = WordCloud(font_path=path, background_color='white', width=500, height=300).generate(expert_comments)

plt.figure(figsize=(15, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

# 특정일 이슈 크롤링

```python
!apt-get update -qq
!apt-get install fonts-nanum* -qq
import requests
import pandas as pd
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import nltk
from konlpy.tag import Kkma
from konlpy.tag import Twitter
from wordcloud import WordCloud

date='20240515'
news_url = 'https://news.naver.com/main/ranking/popularDay.nhn?date={}'.format(date)

# headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)
# AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.90 Safari/537.36'}


req = requests.get(news_url)
```

```python
soup = BeautifulSoup(req.text, 'html.parser')
news_titles = soup.select('.rankingnews_box > ul > li > div > a')
crowled_title = []
for i in range(len(news_titles)):
 crowled_title.append(news_titles[i].text)
 print(i+1, news_titles[i].text)

crowled_title
```

```python
title = "".join(crowled_title)
filtered_title = title.replace('.', ' ').replace('"',' ').replace(',',' ').replace("'","").replace('·', ' ').replace('=',' ').replace('\n',' ')

filtered_title
```

```python
tw = Twitter() # Twitter가 Okt로 변경
tokens_ko = tw.nouns(filtered_title)
tokens_ko
ko = nltk.Text(tokens_ko, name='기사 내 명사')
ko.tokens
ko.vocab()

new_ko=[]
for word in ko:
 if len(word) > 1 and word != '단독' and word != ' ':
  new_ko.append(word)

new_ko
```

```python
ko = nltk.Text(new_ko, name = '기사 내 명사 두 번째')
ko.tokens
ko.vocab()
data = ko.vocab().most_common(150)
data = dict(data)
data
```

```python
wordcloud = WordCloud().generate(filtered_title)
font = '/usr/share/fonts/truetype/nanum/NanumGothicEco.ttf'
wc = WordCloud(font_path=font,\
 background_color="white", \
 width=1000, \
 height=1000, \
 max_words=100, \
 max_font_size=300)
wc = wc.generate_from_frequencies(data)
plt.figure(figsize=(10,10))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
